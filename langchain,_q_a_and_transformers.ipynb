```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline

# Initialize a pre-trained model pipeline from transformers
qa_pipeline = pipeline("question-answering")

# Define a function for answering questions
def answer_question(context, question):
    result = qa_pipeline(question=question, context=context)
    return result['answer']

# Example usage
context = "LangChain makes it easy to build LLM-powered applications by providing modular components and tools."
question = "What can I use langchain for?"
answer = answer_question(context, question)
print("Question:", question)
print("Answer:", answer)
```

```python
import fitz  # PyMuPDF
from transformers import pipeline

# Initialize a pre-trained model pipeline from transformers
qa_pipeline = pipeline("question-answering")

# Function to extract text from a PDF document
def extract_text_from_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        text += page.get_text()
    return text

# Function to answer questions based on the document text
def answer_question_from_document(file_path, question):
    context = extract_text_from_pdf(file_path)
    result = qa_pipeline(question=question, context=context)
    return result['answer']

# Example usage
file_path = "nuro.pdf"
question = "What is nurovant about?"
answer = answer_question_from_document(file_path, question)
print("Question:", question)
print("Answer:", answer)
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Initialize a pre-trained model pipeline from transformers
qa_pipeline = pipeline("question-answering")

# Create a LangChain LLM wrapper
class TransformersLLM(HuggingFacePipeline):
    def __init__(self, pipeline):
        self.pipeline = pipeline

    def generate(self, prompt, max_length=50):
        return self.pipeline(prompt)

# Initialize the LangChain LLM with our pipeline
llm = TransformersLLM(pipeline=qa_pipeline)

# Define a function for answering questions
def answer_question(context, question):
    prompt = {
        "context": context,
        "question": question
    }
    response = llm.generate(prompt)
    return response['answer']

# Example usage
context = "Need advice on what to do about my future. Help by providing expert responses so we can communicate about it"
question = "Is it okay to start music in your late 30s?"
answer = answer_question(context, question)
print("Question:", question)
print("Answer:", answer)
```
